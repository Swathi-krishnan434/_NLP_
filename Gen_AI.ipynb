```python
# Text Generation using GPT-2
from transformers import pipeline

generator = pipeline("text-generation",model="gpt2")

text="Artificial Intelligence is"

result=generator(text,max_length=50,num_return_sequences=1)

print("Input:",text)
print("Generated Text:",result[0]['generated_text'])
```

```python
# Zero-Shot Text Classification
from transformers import pipeline

classifier=pipeline("zero-shot-classification",
model="facebook/bart-large-mnli")

text="The team won the cricket match"

labels=["sports","technology","education"]

result=classifier(text,labels)

print("Text:",text)
print("Label:",result['labels'][0])
```

```python
#  Mask using FillBERT
from transformers import pipeline

model=pipeline("fill-mask",model="bert-base-uncased")

text="Artificial Intelligence is [MASK]"

result=model(text)

for r in result[:5]:
 print(r['sequence'])
```

```python
# Text Similarity using Sentence Transformer

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model=SentenceTransformer("all-MiniLM-L6-v2")

text=["AI is powerful","Artificial Intelligence is useful"]

e=model.encode(text)

print("Similarity:",cosine_similarity([e[0]],[e[1]])[0][0])
```

```python
# Simple RAG Document Retrieval
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

docs=["AI learns from data","Python is a language",
      "Generative AI creates new content"]

model=SentenceTransformer("all-MiniLM-L6-v2")
e=model.encode(docs)

q="what do you mean by generative ai?"
qe=model.encode([q])

i=cosine_similarity(qe,e).argmax()

print("Question:",q)
print("Retrieved:",docs[i])
```

```python
# AI Safety Guardrail
text=input("Enter Prompt: ")

blocked=["hack","attack","steal"]

if any(w in text.lower() for w in blocked):
 print("Unsafe Prompt Detected")
else:
 print("Safe Prompt")
```

```python
# Prompt Injection Detection
prompt=input("Enter Prompt: ")

words=["ignore previous","reveal password","system prompt"]

if any(w in prompt.lower() for w in words):
 print("Prompt Injection Detected")
else:
 print("Safe Prompt")
```

```python
# AI Hallucination Checker
facts=["Python is a programming language",
       "AI means Artificial Intelligence"]

text=input("Enter AI Response: ")

if text in facts:
 print("Verified Information")
else:
 print("Possible Hallucination")
```

```python
# Semantic Search
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

docs=["AI learns from data",
      "Python is used for coding",
      "NLP processes human language"]

model=SentenceTransformer("all-MiniLM-L6-v2")

e=model.encode(docs)
q=model.encode(["language processing"])

score=cosine_similarity(q,e)

print("Best Result:",docs[score.argmax()])
```
